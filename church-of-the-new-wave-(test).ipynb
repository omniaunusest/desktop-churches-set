```python
from PIL import Image
import numpy as np
import os

def distorted_wave(image_path, amplitude, wavelength):
    # Apply a sine wave distortion to the input image.

        #image_path (str): Path to the input image.
        #amplitude (int): Wave's amplitude.
        #wavelength (int): Wave's wavelength.

    #Returns:
        #PIL.Image: Distorted image.

    img = Image.open(image_path)
    img_array = np.array(img)
    height, width, channels = img_array.shape

    # Create coordinate grids
    x = np.tile(np.arange(width), (height, 1))
    y = np.tile(np.arange(height).reshape(-1, 1), (1, width))

    # Apply distortion
    y_distorted = y + amplitude * np.sin(2 * np.pi * x / wavelength)

    # Build the new image
    new_image = np.zeros_like(img_array)
    for i in range(height):
        for j in range(width):
            y_pos = int(y_distorted[i, j]) % height
            new_image[i, j] = img_array[y_pos, j]

    return Image.fromarray(new_image)

def get_image_path(prompt):
    # Prompt the user for an image path, ensuring it exists.
    while True:
        path = input(prompt).strip()
        if os.path.exists(path):
            return path
        print(f"Error: '{path}' not found. Try again.")

def get_int_input(prompt, default=None):
    # Prompt the user for an INT input, with optional default.
    while True:
        try:
            value = input(prompt).strip()
            if not value and default is not None:
                return default
            return int(value)
        except ValueError:
            print("Please, write a valid answer for the new gospel.")

def main(): # aquí falta mucho ASCII y mucha chicha 
    print("=== Church of the New Wave: Distortion's gospel ===")
    image_path = get_image_path("Enter the image path: ")
    amplitude = get_int_input("Enter wave amplitude (e.g., 10): ", 10)
    wavelength = get_int_input("Enter wave wavelength (e.g., 20): ", 20)

    # Apply distortion and save
    output_path = "newwave.png"
    wave_img = distorted_wave(image_path, amplitude, wavelength)
    wave_img.save(output_path)
    print(f"An image has been saved! You can join the distortion here: '{output_path}'")

if __name__ == "__main__":
    main()


# hay que optimizar los inputs (ofrecer cómo llamar)
# hay que implementar strip para manejar extensiones de manera independiente
# hay que implementar ofrecer distorsionar otra imagen
# hay que pensar si aquí procede meter a bailar una API
# hay que plantearse si complejizarlo como para hacer una clase o... mantenerlo sencillo
```
